```python
import torch
from transformers import pipeline

model_id = "NousResearch/Hermes-3-Llama-3.1-8B"

pipe = pipeline(
    "text-generation",
    model=model_id,
    torch_dtype=torch.bfloat16,
    device_map="auto"
)

pipe("The key to life is")
```
